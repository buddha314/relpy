```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
```

```python
np.random.seed(3)
n = 20
x = np.random.uniform(-15, 15, size = n)
y = x**2 + 2*np.random.randn(n, )
X = np.reshape(x ,[n, 1]) 
y = np.reshape(y ,[n ,])
```

```python
y
```

```python
clf = MLPRegressor(alpha=0.001, hidden_layer_sizes = (10,), max_iter = 100000, 
                 activation = 'logistic', verbose = 'False', learning_rate = 'adaptive')
a = clf.fit(X, y)
```

```python
x_ = np.linspace(-4, 4, 160) # define axis

pred_x = np.reshape(x_, [160, 1]) # [160, ] -> [160, 1]
pred_y = clf.predict(pred_x)
```

```python
fig = plt.figure() 
plt.plot(x_, x_**2, color = 'b') # plot original function
plt.plot(pred_x, pred_y, '-') # plot network output
```
